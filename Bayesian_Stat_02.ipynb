```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
import math
from scipy import stats
from scipy.stats import gamma,beta,binom
```

```python
# 3.9

def dGalenshore(y,a,theta):
    prob = (2/math.gamma(a))* (theta**(2*a)) * (y**((2*a) - 1)) * (math.exp(-1*((theta**2)*(y**2))))
    
    return prob

y = np.arange(0.02,5,0.02)

df = pd.DataFrame({'y':y})

df['a1-theta1'] = df['y'].apply(lambda x: dGalenshore(x,1,1))
df['a1-theta3'] = df['y'].apply(lambda x: dGalenshore(x,1,3))
df['a3-theta1'] = df['y'].apply(lambda x: dGalenshore(x,3,1))
df['a2-theta4'] = df['y'].apply(lambda x: dGalenshore(x,2,4))

fig,ax = plt.subplots(1,1,figsize = (5,5))
ax.plot(df['y'],df['a1-theta1'])
ax.plot(df['y'],df['a1-theta3'])
ax.plot(df['y'],df['a3-theta1'])
ax.plot(df['y'],df['a2-theta4'])
ax.legend()
ax.set_title('Members of Galenshore Density',fontsize = 15);
    
    
    
plt.savefig('F:/JAY/COURSES/STAT_638/HomeWork/HW3/3_9(a).png',bbox_inches = 'tight')   
```
